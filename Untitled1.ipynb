```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Define the universe for each input and output
ambient_temp = ctrl.Antecedent(np.arange(0, 40), 'Ambient Temperature')
lighting_level = ctrl.Antecedent(np.arange(0, 1000), 'Lighting Level')
user_activity = ctrl.Antecedent(np.arange(0, 1), 'User Activity')
user_mood = ctrl.Antecedent(np.arange(0, 10), 'User Mood')
humidity_level = ctrl.Antecedent(np.arange(0, 100), 'Humidity level')
time_of_day = ctrl.Antecedent(np.arange(0, 24), 'Time of day')

heater_control = ctrl.Consequent(np.arange(0, 100), 'Heater Control')
cooling_fan_control = ctrl.Consequent(np.arange(0, 100), 'Cooling Fan Control')
blinds_control = ctrl.Consequent(np.arange(0, 100), 'Blinds Control')
dimmer_switch_control = ctrl.Consequent(np.arange(0, 100), 'Dimmer Switch Control')
humidifier_control = ctrl.Consequent(np.arange(0, 1), 'Humidifier')
music_control = ctrl.Consequent(np.arange(0, 100), 'Music')

# Populate the fuzzy membership functions for Ambient Temperature
ambient_temp['Cold'] = fuzz.trimf(ambient_temp.universe, [0, 0, 15])
ambient_temp['Comfortable'] = fuzz.trimf(ambient_temp.universe, [10, 20, 30])
ambient_temp['Hot'] = fuzz.trimf(ambient_temp.universe, [25, 40, 40])

# Lighting Level Membership Functions
lighting_level['Dark'] = fuzz.trimf(lighting_level.universe, [0, 0, 250])
lighting_level['Moderate'] = fuzz.trimf(lighting_level.universe, [100, 500, 900])
lighting_level['Bright'] = fuzz.trimf(lighting_level.universe, [500, 1000, 1000])

# User Activity Membership Functions
user_activity['Not Moving'] = fuzz.trimf(user_activity.universe, [0, 0, 0.5])
user_activity['Moving'] = fuzz.trimf(user_activity.universe, [0.5, 1, 1])

# User Mood Membership Functions
user_mood['Sad'] = fuzz.trimf(user_mood.universe, [0, 0, 5])
user_mood['Neutral'] = fuzz.trimf(user_mood.universe, [3, 5, 7])
user_mood['Happy'] = fuzz.trimf(user_mood.universe, [5, 10, 10])

# Humidity Level Membership Functions
humidity_level['Dry'] = fuzz.trimf(humidity_level.universe, [0, 0, 40])
humidity_level['Moderate'] = fuzz.trimf(humidity_level.universe, [30, 50, 70])
humidity_level['Humid'] = fuzz.trimf(humidity_level.universe, [60, 100, 100])

# Time of Day Membership Functions
time_of_day['Morning'] = fuzz.trimf(time_of_day.universe, [0, 6, 12])
time_of_day['Afternoon'] = fuzz.trimf(time_of_day.universe, [12, 14, 16])
time_of_day['Evening'] = fuzz.trimf(time_of_day.universe, [16, 18, 20])
time_of_day['Night'] = fuzz.trimf(time_of_day.universe, [20, 22, 24])

# OUTPUTS.

# Populate the fuzzy membership functions for Heater Control
heater_control['Low'] = fuzz.trimf(heater_control.universe, [0, 0, 33])
heater_control['Medium'] = fuzz.trimf(heater_control.universe, [0, 33, 66])
heater_control['High'] = fuzz.trimf(heater_control.universe, [33, 100, 100])

# Cooling Fan Control Membership Functions
cooling_fan_control['Low'] = fuzz.trimf(cooling_fan_control.universe, [0, 0, 33])
cooling_fan_control['Medium'] = fuzz.trimf(cooling_fan_control.universe, [0, 33, 66])
cooling_fan_control['High'] = fuzz.trimf(cooling_fan_control.universe, [66, 100, 100])

# Blinds Control Membership Functions
blinds_control['Closed'] = fuzz.trimf(blinds_control.universe, [0, 0, 33])
blinds_control['Partially opened'] = fuzz.trimf(blinds_control.universe, [0, 33, 66])
blinds_control['Fully open'] = fuzz.trimf(blinds_control.universe, [66, 100, 100])

# Dimmer Switch Control Membership Functions
dimmer_switch_control['Low'] = fuzz.trimf(dimmer_switch_control.universe, [0, 0, 33])
dimmer_switch_control['Medium'] = fuzz.trimf(dimmer_switch_control.universe, [0, 33, 66])
dimmer_switch_control['High'] = fuzz.trimf(dimmer_switch_control.universe, [66, 100, 100])

# Humidifier Control Membership Functions
humidifier_control['Off'] = fuzz.trimf(humidifier_control.universe, [0, 0.5, 1])
humidifier_control['On'] = fuzz.trimf(humidifier_control.universe, [0.5, 1, 1])

# Music Control Membership Functions
music_control['Off'] = fuzz.trimf(music_control.universe, [0, 18, 30])
music_control['Soft'] = fuzz.trimf(music_control.universe, [30, 50, 70])
music_control['High'] = fuzz.trimf(music_control.universe, [75, 100, 100])

# RULES.

# Define fuzzy rules
rule1 = ctrl.Rule(antecedent=(time_of_day['Morning'] & ambient_temp['Cold']),
                  consequent=(heater_control['High'], music_control['Soft']))

rule2 = ctrl.Rule(antecedent=(time_of_day['Morning'] & lighting_level['Bright']),
                  consequent=(blinds_control['Partially opened'], dimmer_switch_control['Low'], music_control['Off']))

rule3 = ctrl.Rule(antecedent=(time_of_day['Afternoon'] & user_mood['Happy']),
                  consequent=(cooling_fan_control['Low'], dimmer_switch_control['Low'], music_control['Soft']))

rule4 = ctrl.Rule(antecedent=(time_of_day['Evening'] & user_mood['Neutral']),
                  consequent=(blinds_control['Closed'], dimmer_switch_control['High'], music_control['Soft']))

rule5 = ctrl.Rule(antecedent=(humidity_level['Humid']),
                  consequent=(humidifier_control['Off'], cooling_fan_control['High']))

rule6 = ctrl.Rule(antecedent=(user_mood['Neutral'] & lighting_level['Moderate']),
                  consequent=(blinds_control['Partially opened'], dimmer_switch_control['High']))

rule7 = ctrl.Rule(antecedent=(time_of_day['Night'] & ambient_temp['Comfortable']),
                  consequent=(heater_control['Low'], blinds_control['Closed'], music_control['Off']))

rule8 = ctrl.Rule(antecedent=(user_activity['Moving'] & ambient_temp['Hot']),
                  consequent=(cooling_fan_control['High']))

rule9 = ctrl.Rule(antecedent=(time_of_day['Morning'] & humidity_level['Dry']),
                  consequent=(humidifier_control['On']))

rule10 = ctrl.Rule(antecedent=(ambient_temp['Hot']) & time_of_day['Afternoon'],
                   consequent=(blinds_control['Partially opened'], cooling_fan_control['Medium']))

rule11 = ctrl.Rule(antecedent=(humidity_level['Humid'] & ambient_temp['Cold']),
                   consequent=(heater_control['High'], humidifier_control['Off']))

rule12 = ctrl.Rule(antecedent=(time_of_day['Afternoon'] & humidity_level['Humid']),
                   consequent=(cooling_fan_control['High'], humidifier_control['Off']))

rule13 = ctrl.Rule(antecedent=(time_of_day['Evening'] & user_mood['Sad']),
                   consequent=(dimmer_switch_control['Low'], music_control['Soft']))

rule14 = ctrl.Rule(antecedent=(time_of_day['Morning'] & humidity_level['Moderate']),
                   consequent=(blinds_control['Fully open'], humidifier_control['On']))

rule15 = ctrl.Rule(antecedent=(ambient_temp['Cold']),
                   consequent=(blinds_control['Partially opened'], heater_control['Medium'], dimmer_switch_control['Low']))

rule16 = ctrl.Rule(antecedent=(user_activity['Moving'] & ambient_temp['Hot']),
                   consequent=(cooling_fan_control['High']))

rule17 = ctrl.Rule(antecedent=(time_of_day['Night'] & user_mood['Happy']),
                   consequent=(music_control['Soft'], dimmer_switch_control['Low']))

rule18 = ctrl.Rule(antecedent=(time_of_day['Night'] & humidity_level['Dry']),
                   consequent=(humidifier_control['On']))

rule19 = ctrl.Rule(antecedent=(time_of_day['Morning'] & user_activity['Moving']),
                   consequent=(blinds_control['Fully open'], music_control['High']))


rule20 = ctrl.Rule(antecedent=(ambient_temp['Cold']),
                   consequent=(blinds_control['Partially opened'], dimmer_switch_control['High']))

rule21 = ctrl.Rule(antecedent=(time_of_day['Evening'] & user_activity['Not Moving']),
                   consequent=(blinds_control['Closed'], dimmer_switch_control['Low'], music_control['Soft']))

rule22 = ctrl.Rule(antecedent=(time_of_day['Afternoon'] & user_mood['Neutral']),
                   consequent=(blinds_control['Partially opened'], dimmer_switch_control['Low']))

rule23 = ctrl.Rule(antecedent=(time_of_day['Morning'] & ambient_temp['Hot']),
                   consequent=(cooling_fan_control['Low']))

rule24 = ctrl.Rule(antecedent=(humidity_level['Moderate']),
                   consequent=(humidifier_control['Off']))

rule25 = ctrl.Rule(antecedent=(ambient_temp['Cold'] & humidity_level['Dry']),
                  consequent=humidifier_control['On'])

rule26 = ctrl.Rule(antecedent=(ambient_temp['Hot'] & humidity_level['Moderate']),
                  consequent=humidifier_control['Off'])

rule27 = ctrl.Rule(antecedent=(ambient_temp['Hot'] & humidity_level['Humid']),
                  consequent=humidifier_control['Off'])

rule28 = ctrl.Rule(antecedent=(ambient_temp['Cold'] & humidity_level['Humid']),
                  consequent=humidifier_control['Off'])

rule29 = ctrl.Rule(antecedent=(ambient_temp['Cold'] & humidity_level['Dry'] & time_of_day['Morning']),
                  consequent=humidifier_control['On'])

# ... create all other rules following the structure given in your FIS file

# Define the control system
heating_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10,
             rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19,
             rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27, rule28, rule29])
heating = ctrl.ControlSystemSimulation(heating_ctrl)



# Setting input values for the fuzzy control system simulation
heating.input['Ambient Temperature'] = 22
heating.input['Lighting Level'] = 250
heating.input['User Activity'] = 1
heating.input['User Mood'] = 7
heating.input['Humidity level'] = 60
heating.input['Time of day'] = 10

# Execute the computation
heating.compute()

# Accessing and printing the outputs
print("Heater Control:", heating.output['Heater Control'])
print("Cooling Fan Control:", heating.output['Cooling Fan Control'])
print("Blinds Control:", heating.output['Blinds Control'])
print("Dimmer Switch Control:", heating.output['Dimmer Switch Control'])
print("Humidifier Control:", heating.output['Humidifier Control'])
print("Music Control:", heating.output['Music Control'])


```
